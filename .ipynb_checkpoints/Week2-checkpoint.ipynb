```python
######################例6-4-9
from tkinter import *

def sele():
    seleItem="你选了选项"+str(rVar.get())
    lbl.config(text=seleItem)
    
window=Tk()
window.title("单选按钮")

rVar=IntVar() ##整型变量
for i in range(3):
    r=Radiobutton(window,text="选项"+str(i+1),variable=rVar,
                  value=i+1,command=sele)
    r.grid()
lbl=Label(window,text="你尚未选择！",width=36)
lbl.grid()

window.mainloop()
```

```python
######################例7-2-10
from tkinter import *

def sele():
    showStr="你选了"
    checkVar=checkVar1.get()+checkVar2.get()
    if checkVar==1:
        showStr+="滑雪！"
    elif checkVar==2:
        showStr+="射击！"
    elif checkVar==3:
        showStr+="滑雪和射击！"
    else:
        showStr="你去掉了所有选项！"
    lbl.config(text=showStr)

window=Tk()
window.title("复选按钮")

checkVar1=IntVar()
checkVar2=IntVar()
c1=Checkbutton(window,text="滑雪",variable=checkVar1,
              onvalue=1,offvalue=0,command=sele)
c2=Checkbutton(window,text="射击",variable=checkVar2,
              onvalue=2,offvalue=0,command=sele)
c1.grid()
c2.grid()

lbl=Label(window,text="你尚未选择！",width=36)
lbl.grid()
window.mainloop()
```

```python
######################例7-2-11
from tkinter import *
from tkinter import ttk
window = Tk() 
window.title("下拉列表")

lbl=Label(window,text="请从下拉表中选择",width=50)
lbl.grid()

value=StringVar()
cbox=ttk.Combobox(window,width=10,textvariable=value)
cbox['values']=('高中','大专','本科','硕士','博士')
cbox.grid()
cbox.current(0) #设置下拉列表初次显示的项

def show_msg(event):
    lbl.config(text='你的选择是：'+cbox.get())
cbox.bind("<<ComboboxSelected>>",show_msg)

window.mainloop()
```

```python
######################例7-2-12
from tkinter import *

window=Tk()
window.geometry('500x300')
window.title("下拉菜单")

def openCall():
    print('你点击了Open选项')
def saveCall():
    print('你点击了Save选项')
def aboutCall():
    print('这是菜单的简单演示')
    
menubar=Menu(window)

filemenu=Menu(menubar，tearoff=False) #file下拉菜单 tearoff下拉菜单不可独立
filemenu.add_command(label='Open',command=openCall)
filemenu.add_command(label='Save',command=saveCall)
filemenu.add_separator() #分割线
filemenu.add_command(label='Exit',command=window.destroy)
menubar.add_cascade(label='File',menu=filemenu) #加入菜单栏

helpmenu=Menu(menubar) #help下拉菜单
helpmenu.add_command(label='About',command=aboutCall)
menubar.add_cascade(label='Help',menu=helpmenu) #加入菜单栏

window.config(menu=menubar)
window.mainloop()
```

```python
######################9.2.3
from tkinter import *
window = Tk() 

#font=(" ", 12, "bold") 有特定格式？？？
def sele(): 
    txt.tag_add('myArea',1.0,END)
    checkVar=checkVar1.get()+checkVar2.get()
    if checkVar==1:
        txt.tag_config('myArea',foreground=colVar.get(),font=(" ", 12, 'bold'))
    elif checkVar==2:
        txt.tag_config('myArea',foreground=colVar.get(),font=(" ", 12, 'italic'))
    elif checkVar==3:
        txt.tag_config('myArea',foreground=colVar.get(),font=(" ", 12, 'bold italic'))
    else:
        txt.tag_config('myArea',foreground=colVar.get(),font=(" ", 12))    
    
#多行文本
txt=Text(window,height=5,width=50)
txt.insert(1.0,'此处编辑你的文本！')
txt.grid(row=0,column=0,columnspan=5,sticky=W+E)

#设置框架
frame=Frame(window)
frame.grid()

#单选按钮
colVar=StringVar()
r1=Radiobutton(frame,text="红色",variable=colVar,
              value='red',command=sele)
r2=Radiobutton(frame,text="绿色",variable=colVar,
              value='green',command=sele)
r3=Radiobutton(frame,text="蓝色",variable=colVar,
              value='blue',command=sele)
r1.grid(row=0,column=0)
r2.grid(row=0,column=1)
r3.grid(row=0,column=2)

#复选按钮
checkVar1=IntVar()
checkVar2=IntVar()
c1=Checkbutton(frame,text="粗体",variable=checkVar1,
              onvalue=1,offvalue=0,command=sele)
c2=Checkbutton(frame,text="斜体",variable=checkVar2,
              onvalue=2,offvalue=0,command=sele)
c1.grid(row=0,column=3)
c2.grid(row=0,column=4)

window.mainloop()
```

```python
######################9.2.5
from tkinter import *
window = Tk() 

def showEvent1(event):
    cv.delete(ALL)
    cv.create_text(250,10,text='x='+str(event.x)+'y='+str(event.y))
    cv.create_line(event.x,event.y-5,event.x,event.y+5,width=1)
    cv.create_line(event.x-5,event.y,event.x+5,event.y,width=1)
def showEvent2(event):
    cv.delete(ALL)

#左击和双击的数字都是1
window.bind('<Button-1>',showEvent1) 
window.bind('<Double-Button-1>',showEvent2)

cv=Canvas(window,height=500,width=500)
cv.pack()

window.mainloop()
```

## d.案例应用

```python
from tkinter import *
from tkinter.messagebox import showinfo

class Account:
    def __init__(self,number="",money=0):
        self.id=number         #为id赋予账号(number)
        self.balance=money
    def resetAccount(self,number,money):
            self.id=number
            self.balance=money
    def deposit(self,money):
        self.balance+=money
    def withdraw(self,money):
        if money>self.balance:
            print ("账户余额不足，本次取款操作失败！")
        else:
            self.balance-=money
    def display(self):
        print ("账户：%s 的当前账户余额是：%.2f元。"%(self.id,self.balance))

def openCall():
    name=entry_name.get()
    money=entry_inmoney.get()
    if name=="":
            showinfo(title='错误提示', message='开户必须输入账号')
            return 
    if money=="":
            money=0
    else:
            money=float(money)
    a1.resetAccount(name,money)
    showinfo(title='开户成功提示', message='已成功开户！')
    a1.display()
def displayCall():
    a1.display()
def aboutCall():
    print('这是银行账户管理')

def depositCall():
    money=float(entry_inmoney.get())
    a1.deposit(money)
    showinfo(title='存款成功', message='已存入'+str(money)+'元')
    a1.display()
def withdrawCall():
    money=float(entry_outmoney.get())
    a1.withdraw(money)
    showinfo(title='取款成功', message='已取出'+str(money)+'元')
    a1.display()
    
window = Tk()
window.geometry("350x320")
window.title("银行账户管理")

#创建菜单栏menubar
menubar = Menu(window)                                   
#创建下拉菜单File，然后将其加入到顶级的菜单栏menubar中
filemenu = Menu(menubar,tearoff=0)                            #生成下拉菜单
#在下拉菜单中添加各项菜单项（命令）
filemenu.add_command(label="开户", command=openCall) 
filemenu.add_command(label="查看", command=displayCall)
filemenu.add_separator()                        #添加菜单项分割线
filemenu.add_command(label="退出", command=window.destroy)
menubar.add_cascade(label="业务", menu=filemenu)   #将下拉菜单增加到菜单栏中
#创建另一个下拉菜单Help
helpmenu = Menu(menubar,tearoff=0)
helpmenu.add_command(label="About", command=aboutCall)
menubar.add_cascade(label="帮助", menu=helpmenu)
window.config(menu=menubar)

##第1行
lb_name=Label(window,text='账号',)
lb_name.grid(row=0,column=0,ipadx=10,ipady=10,padx=10,pady=10)
entry_name = Entry(window,width=20)
entry_name.grid(row=0,column=1,ipadx=20,ipady=10,padx=10,pady=10)
##第2行
lb_inmoney=Label(window,text='存入金额',)
lb_inmoney.grid(row=1,column=0,ipadx=10,ipady=10,padx=10,pady=10)
entry_inmoney = Entry(window,width=20)
entry_inmoney.grid(row=1,column=1,ipadx=20,ipady=10,padx=10,pady=10)
##第3行
lb_outmoney=Label(window,text='取出金额',)
lb_outmoney.grid(row=2,column=0,ipadx=10,ipady=10,padx=10,pady=10)
entry_outmoney = Entry(window,width=20)
entry_outmoney.grid(row=2,column=1,ipadx=20,ipady=10,padx=10,pady=10)
#第4行
btn_deposit=Button(window,text="存款",width=10,command=depositCall)
btn_deposit.grid(row=3,column=0,ipadx=10,ipady=10,padx=10,pady=10)
btn_withdraw=Button(window,text="取款",width=10,command=withdrawCall)
btn_withdraw.grid(row=3,column=1,ipadx=10,ipady=10,padx=10,pady=10)
    
a1=Account()

window.mainloop()
```
